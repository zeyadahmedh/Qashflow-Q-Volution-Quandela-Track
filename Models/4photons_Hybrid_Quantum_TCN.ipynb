```python
"""
Quantum Machine Learning for Option Pricing
Fixed version with explicit device handling for quantum layer outputs
"""

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tqdm import tqdm
import math
import os
from datetime import datetime

# Import only what we know exists from documentation
from merlin import (
    QuantumLayer,
    MeasurementStrategy,
    ComputationSpace,
    LexGrouping,
    ModGrouping
)
import perceval as pcvl

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)

class ScaleLayer(nn.Module):
    """
    Scale input values to appropriate range for quantum circuit
    """
    def __init__(self, dim, scale_type="learned"):
        super().__init__()
        if scale_type == "learned":
            self.scale = nn.Parameter(torch.full((dim,), 0.1))
        elif scale_type == "pi":
            self.scale = torch.full((dim,), torch.pi)
        elif scale_type == "2pi":
            self.scale = torch.full((dim,), 2 * torch.pi)
        else:
            self.scale = torch.full((dim,), 1.0)

    def forward(self, x):
        return x * self.scale

class Chomp1d(nn.Module):
    """
    Chomp1d module for causal convolutions
    Removes excess padding from the right side
    """
    def __init__(self, chomp_size):
        super().__init__()
        self.chomp_size = chomp_size

    def forward(self, x):
        return x[:, :, :-self.chomp_size].contiguous()

class TemporalBlock(nn.Module):
    """
    Temporal Block for TCN
    Consists of two dilated causal convolutions with residual connection
    """
    def __init__(self, n_inputs, n_outputs, kernel_size, stride, dilation, padding, dropout=0.2):
        super().__init__()

        # First convolution
        self.conv1 = nn.Conv1d(
            n_inputs, n_outputs, kernel_size,
            stride=stride, padding=padding, dilation=dilation
        )
        self.chomp1 = Chomp1d(padding)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(dropout)

        # Second convolution
        self.conv2 = nn.Conv1d(
            n_outputs, n_outputs, kernel_size,
            stride=stride, padding=padding, dilation=dilation
        )
        self.chomp2 = Chomp1d(padding)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(dropout)

        # Residual connection
        self.net = nn.Sequential(
            self.conv1, self.chomp1, self.relu1, self.dropout1,
            self.conv2, self.chomp2, self.relu2, self.dropout2
        )

        # Downsample if input and output dimensions differ
        self.downsample = nn.Conv1d(n_inputs, n_outputs, 1) if n_inputs != n_outputs else None
        self.relu = nn.ReLU()

    def forward(self, x):
        out = self.net(x)
        res = x if self.downsample is None else self.downsample(x)
        return self.relu(out + res)

class TemporalConvNet(nn.Module):
    """
    Temporal Convolutional Network
    """
    def __init__(self, num_inputs, num_channels, kernel_size=3, dropout=0.2):
        super().__init__()

        layers = []
        num_levels = len(num_channels)

        for i in range(num_levels):
            dilation_size = 2 ** i
            in_channels = num_inputs if i == 0 else num_channels[i-1]
            out_channels = num_channels[i]

            layers.append(
                TemporalBlock(
                    in_channels, out_channels, kernel_size,
                    stride=1, dilation=dilation_size,
                    padding=(kernel_size - 1) * dilation_size,
                    dropout=dropout
                )
            )

        self.network = nn.Sequential(*layers)

    def forward(self, x):
        """
        Args:
            x: [batch_size, channels, sequence_length]
        Returns:
            [batch_size, channels, sequence_length]
        """
        return self.network(x)

def create_circuit_general(modes=6):
    """
    Create a variational quantum circuit with specified number of modes
    """
    # Left interferometer with trainable parameters
    wl = pcvl.GenericInterferometer(
        modes,
        lambda i: (
            pcvl.BS(pcvl.P(f"theta_li{i}_bs"))
            // pcvl.PS(pcvl.P(f"theta_li_ps{i}"))
            // pcvl.BS(pcvl.P(f"theta_lo{i}_bs"))
            // pcvl.PS(pcvl.P(f"theta_lo_ps{i}"))
        ),
        shape=pcvl.InterferometerShape.RECTANGLE,
    )

    # Variable phase shifters for input encoding
    c_var = pcvl.Circuit(modes)
    # We'll encode features in all modes
    for i in range(modes):
        px = pcvl.P(f"px{i}")
        c_var.add(i, pcvl.PS(px))

    # Right interferometer with trainable parameters
    wr = pcvl.GenericInterferometer(
        modes,
        lambda i: (
            pcvl.BS(pcvl.P(f"theta_ri{i}_bs"))
            // pcvl.PS(pcvl.P(f"theta_ri{i}_ps"))
            // pcvl.BS(pcvl.P(f"theta_ro{i}_bs"))
            // pcvl.PS(pcvl.P(f"theta_ro{i}_ps"))
        ),
        shape=pcvl.InterferometerShape.RECTANGLE,
    )

    # Combine all components
    c = pcvl.Circuit(modes)
    c.add(0, wl, merge=True)
    c.add(0, c_var, merge=True)
    c.add(0, wr, merge=True)

    return c

class QuantumFinancialTCN(nn.Module):
    """
    Hybrid Quantum-Classical Model for Option Pricing using Temporal Convolutional Network
    """

    def __init__(self, n_features=200, q_modes=6, n_photons=4,  # Changed to 4 photons
                 hidden_channels=[64, 128, 64], kernel_size=3, dropout=0.1):
        super().__init__()

        # Store dimensions
        self.n_features = n_features
        self.q_modes = q_modes
        self.n_photons = n_photons  # Store photon count
        self.hidden_channels = hidden_channels
        self.kernel_size = kernel_size
        self.dropout = dropout

        # Step 1: Classical compression (n_features -> q_modes)
        self.compressor = nn.Sequential(
            nn.Linear(n_features, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, q_modes),
            nn.Tanh()  # Bound to [-1, 1] for phase encoders
        )

        # Step 2: Scale layer
        self.scale_layer = ScaleLayer(q_modes, scale_type="learned")

        # Step 3: Create quantum circuit
        print(f"Creating quantum circuit with {q_modes} modes, {n_photons} photons...")
        circuit = create_circuit_general(modes=q_modes)

        # Create input state based on photon distribution
        input_state = [1] * n_photons + [0] * (q_modes - n_photons)
        print(f"Input state: {input_state}")

        # Create quantum layer
        self.quantum_layer = QuantumLayer(
            input_size=q_modes,
            circuit=circuit,
            trainable_parameters=["theta"],
            input_parameters=["px"],
            input_state=input_state,
            measurement_strategy=MeasurementStrategy.probs(
                computation_space=ComputationSpace.FOCK
            )
        )

        # Calculate quantum output size
        from math import comb
        self.quantum_output_size = comb(q_modes + n_photons - 1, n_photons)
        print(f"Quantum output size: {self.quantum_output_size}")

        # Step 4: Combine classical and quantum features
        self.combined_size = n_features + self.quantum_output_size

        # Step 5: Temporal Convolutional Network
        self.tcn = TemporalConvNet(
            num_inputs=self.combined_size,
            num_channels=hidden_channels,
            kernel_size=kernel_size,
            dropout=dropout
        )

        # Step 6: Output projection
        self.output_projection = nn.Sequential(
            nn.Linear(hidden_channels[-1], hidden_channels[-1] // 2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_channels[-1] // 2, n_features)
        )

        # Initialize parameters
        self._init_weights()

        # Device will be set during first forward pass
        self.device = None

    def _init_weights(self):
        """Initialize weights using Xavier initialization"""
        for module in self.modules():
            if isinstance(module, nn.Linear):
                nn.init.xavier_uniform_(module.weight)
                if module.bias is not None:
                    nn.init.constant_(module.bias, 0)
            elif isinstance(module, nn.Conv1d):
                nn.init.kaiming_normal_(module.weight, mode='fan_in', nonlinearity='relu')
                if module.bias is not None:
                    nn.init.constant_(module.bias, 0)

    def to(self, device):
        """Override to method to track device"""
        self.device = device
        return super().to(device)

    def forward(self, x):
        """
        Forward pass
        Args:
            x: Input tensor of shape [batch_size, seq_len, n_features]
        Returns:
            Predictions of shape [batch_size, seq_len, n_features]
        """
        # Track device
        if self.device is None:
            self.device = x.device

        batch_size, seq_len, _ = x.shape

        # Process each time step through quantum layer
        quantum_features = []
        for t in range(seq_len):
            # Get features for this time step
            x_t = x[:, t, :]  # [batch_size, n_features]

            # Compress features
            x_compressed = self.compressor(x_t)  # [batch_size, q_modes]
            x_scaled = self.scale_layer(x_compressed)

            # Get quantum features
            q_out = self.quantum_layer(x_scaled)  # [batch_size, quantum_output_size]

            # IMPORTANT: Ensure quantum output is on the same device as input
            q_out = q_out.to(self.device)

            quantum_features.append(q_out)

        # Stack quantum features
        quantum_stack = torch.stack(quantum_features, dim=1)  # [batch_size, seq_len, quantum_output_size]

        # Ensure quantum stack is on the same device as input
        quantum_stack = quantum_stack.to(self.device)

        # Combine classical and quantum features
        combined = torch.cat([x, quantum_stack], dim=-1)  # [batch_size, seq_len, combined_size]

        # TCN expects [batch_size, channels, seq_len]
        tcn_input = combined.transpose(1, 2)  # [batch_size, combined_size, seq_len]

        # Temporal convolution processing
        tcn_output = self.tcn(tcn_input)  # [batch_size, hidden_channels[-1], seq_len]

        # Transpose back to [batch_size, seq_len, hidden_channels[-1]]
        tcn_output = tcn_output.transpose(1, 2)  # [batch_size, seq_len, hidden_channels[-1]]

        # Generate predictions
        predictions = self.output_projection(tcn_output)  # [batch_size, seq_len, n_features]

        return predictions

def load_actual_data():
    """
    Load the actual dataset from the hackathon
    """
    try:
        import aqora_cli
        import polars as pl
        from aqora_cli.pyarrow import dataset

        print("Loading data from aqora_cli...")
        df_raw = pl.scan_pyarrow_dataset(dataset("quandela/challenge-swpations", "v0.0.0")).collect()
        df = df_raw.to_pandas()

        # Handle date column
        if 'Date' in df.columns:
            dates = df['Date']
            df = df.drop(columns=['Date'])

        print(f"Data loaded successfully! Shape: {df.shape}")
        print(f"Number of features: {df.shape[1]}")

        return df, dates if 'dates' in locals() else None

    except ImportError as e:
        print(f"Could not load actual data: {e}")
        print("Creating synthetic data for testing...")

        np.random.seed(42)
        n_samples = 200
        n_features = 200

        data = np.zeros((n_samples, n_features))

        # Generate prices with some correlation structure
        for i in range(1, n_samples):
            returns = np.random.randn(n_features) * 0.02
            data[i] = data[i-1] * 0.95 + returns * 0.05 + 0.0001

        # Ensure positivity
        data = np.abs(data) + 0.01

        df = pd.DataFrame(data, columns=[f'feature_{i}' for i in range(n_features)])

        print(f"Created sample data with shape: {df.shape}")
        return df, None

def prepare_data(df, sequence_length=10, test_size=0.2):
    """
    Prepare time series data for training
    """
    if isinstance(df, pd.DataFrame):
        data = df.values.astype(np.float32)
    else:
        data = df.astype(np.float32)

    print(f"Data shape: {data.shape}")
    print(f"Number of features: {data.shape[1]}")

    # Normalize the data
    scaler = StandardScaler()
    data_normalized = scaler.fit_transform(data)

    # Create sequences
    X, y = [], []
    for i in range(len(data_normalized) - sequence_length):
        X.append(data_normalized[i:i+sequence_length])
        y.append(data_normalized[i+sequence_length])

    X = np.array(X, dtype=np.float32)
    y = np.array(y, dtype=np.float32)

    print(f"Created {len(X)} sequences of length {sequence_length}")
    print(f"X shape: {X.shape}, y shape: {y.shape}")

    split_idx = int(len(X) * (1 - test_size))
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]

    X_train = torch.FloatTensor(X_train)
    y_train = torch.FloatTensor(y_train)
    X_test = torch.FloatTensor(X_test)
    y_test = torch.FloatTensor(y_test)

    print(f"Training set: X {X_train.shape}, y {y_train.shape}")
    print(f"Test set: X {X_test.shape}, y {y_test.shape}")

    return X_train, y_train, X_test, y_test, scaler

def train_model(model, train_loader, val_loader, epochs=10, learning_rate=0.001):  # Changed epochs to 10
    """
    Train the hybrid quantum-classical model
    """
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    print(f"Training on device: {device}")

    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.5)

    train_losses = []
    val_losses = []

    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss = 0
        train_progress = tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs} [Train]', leave=False)

        for batch_X, batch_y in train_progress:
            # Move data to device
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)

            optimizer.zero_grad()
            outputs = model(batch_X)

            # For prediction, we want to predict the next time step
            loss = criterion(outputs[:, -1, :], batch_y)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            train_loss += loss.item()
            train_progress.set_postfix({'loss': loss.item()})

        # Validation phase
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X = batch_X.to(device)
                batch_y = batch_y.to(device)
                outputs = model(batch_X)
                loss = criterion(outputs[:, -1, :], batch_y)
                val_loss += loss.item()

        # Calculate average losses
        train_loss /= len(train_loader)
        val_loss /= len(val_loader)
        train_losses.append(train_loss)
        val_losses.append(val_loss)

        # Update learning rate
        scheduler.step(val_loss)

        if (epoch + 1) % 2 == 0:  # Print every 2 epochs
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}')

    return train_losses, val_losses

def save_model(model, scaler, train_losses, val_losses, model_name="quantum_tcn_option_pricing_4photons"):  # Updated name
    """
    Save the trained model and associated data
    """
    # Create models directory if it doesn't exist
    if not os.path.exists('saved_models'):
        os.makedirs('saved_models')

    # Generate timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    # Save model state dict
    model_path = f'saved_models/{model_name}_{timestamp}.pth'
    torch.save({
        'model_state_dict': model.state_dict(),
        'model_config': {
            'n_features': model.n_features,
            'q_modes': model.q_modes,
            'n_photons': model.n_photons,  # Now using the stored photon count
            'hidden_channels': model.hidden_channels,
            'kernel_size': model.kernel_size,
            'dropout': model.dropout
        },
        'train_losses': train_losses,
        'val_losses': val_losses,
        'timestamp': timestamp
    }, model_path)

    # Save scaler separately
    import joblib
    scaler_path = f'saved_models/scaler_{timestamp}.pkl'
    joblib.dump(scaler, scaler_path)

    print(f"\nModel saved to: {model_path}")
    print(f"Scaler saved to: {scaler_path}")

    return model_path, scaler_path

def main():
    """
    Main execution function
    """
    print("=" * 60)
    print("QUANTUM MACHINE LEARNING FOR OPTION PRICING (TCN VERSION WITH 4 PHOTONS)")  # Updated title
    print("=" * 60)

    # Step 1: Load data
    print("\n1. LOADING DATA")
    print("-" * 40)
    df, dates = load_actual_data()

    # Step 2: Prepare data
    print("\n2. PREPARING DATA")
    print("-" * 40)
    X_train, y_train, X_test, y_test, scaler = prepare_data(
        df, sequence_length=10, test_size=0.2
    )

    # Create data loaders
    train_dataset = TensorDataset(X_train, y_train)
    test_dataset = TensorDataset(X_test, y_test)

    train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)

    print(f"Training samples: {len(X_train)}")
    print(f"Testing samples: {len(X_test)}")
    print(f"Sequence length: {X_train.shape[1]}")
    print(f"Features per time step: {X_train.shape[2]}")

    # Step 3: Initialize model
    print("\n3. INITIALIZING QUANTUM-CLASSICAL TCN WITH 4 PHOTONS")  # Updated
    print("-" * 40)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    model = QuantumFinancialTCN(
        n_features=X_train.shape[2],
        q_modes=6,
        n_photons=4,  # Changed to 4 photons
        hidden_channels=[64, 128, 64],  # TCN hidden channels
        kernel_size=3,                    # Kernel size for convolutions
        dropout=0.1                        # Dropout rate
    )
    model = model.to(device)

    total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total trainable parameters: {total_params:,}")

    print("\nTesting forward pass...")
    test_batch = next(iter(train_loader))[0][:2].to(device)
    with torch.no_grad():
        test_output = model(test_batch)
        print(f"Test input shape: {test_batch.shape}")
        print(f"Test output shape: {test_output.shape}")
        print(f"Test output range: [{test_output.min():.4f}, {test_output.max():.4f}]")

    print("\n4. TRAINING MODEL (10 EPOCHS)")  # Updated
    print("-" * 40)
    train_losses, val_losses = train_model(
        model, train_loader, test_loader,
        epochs=10, learning_rate=0.001  # Set to 10 epochs
    )

    print("\n5. VISUALIZING RESULTS")
    print("-" * 40)

    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    axes[0].plot(train_losses, label='Training Loss', linewidth=2)
    axes[0].plot(val_losses, label='Validation Loss', linewidth=2)
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('MSE Loss')
    axes[0].set_title('Training and Validation Loss (TCN with 4 Photons)')  # Updated
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    axes[0].set_yscale('log')

    # Plot sample predictions
    model.eval()
    with torch.no_grad():
        # Get a batch and move to device
        sample_X, sample_y = next(iter(test_loader))
        sample_X = sample_X.to(device)
        sample_y = sample_y.to(device)

        predictions = model(sample_X)

        pred_last = predictions[:, -1, :].cpu()  # Move back to CPU for plotting
        sample_y_cpu = sample_y.cpu()

        # Plot first 10 features for first 3 samples
        n_samples_plot = min(3, len(sample_X))
        n_features_plot = min(10, sample_X.shape[2])

        for i in range(n_samples_plot):
            axes[1].plot(range(n_features_plot),
                        sample_y_cpu[i, :n_features_plot].numpy(),
                        'o-', label=f'Actual {i+1}', alpha=0.7, markersize=4)
            axes[1].plot(range(n_features_plot),
                        pred_last[i, :n_features_plot].numpy(),
                        'x--', label=f'Pred {i+1}', alpha=0.7, markersize=4)

        axes[1].set_xlabel('Feature Index')
        axes[1].set_ylabel('Normalized Price')
        axes[1].set_title('Sample Predictions vs Actual (Next Time Step)')
        axes[1].legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8)
        axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('option_pricing_tcn_results_4photons.png', dpi=150, bbox_inches='tight')  # Updated filename
    plt.show()

    # Step 6: Final evaluation
    print("\n6. FINAL EVALUATION")
    print("-" * 40)
    model.eval()
    with torch.no_grad():
        test_loss = 0
        all_predictions = []
        all_actuals = []

        for batch_X, batch_y in test_loader:
            batch_X = batch_X.to(device)
            batch_y = batch_y.to(device)

            outputs = model(batch_X)
            pred_last = outputs[:, -1, :]
            loss = nn.MSELoss()(pred_last, batch_y)
            test_loss += loss.item()

            all_predictions.append(pred_last.cpu().numpy())
            all_actuals.append(batch_y.cpu().numpy())

        test_loss /= len(test_loader)

        # Calculate additional metrics
        all_predictions = np.concatenate(all_predictions, axis=0)
        all_actuals = np.concatenate(all_actuals, axis=0)

        mae = np.mean(np.abs(all_predictions - all_actuals))
        mape = np.mean(np.abs((all_predictions - all_actuals) / (all_actuals + 1e-8))) * 100

        print(f"Test MSE:  {test_loss:.6f}")
        print(f"Test MAE:  {mae:.6f}")
        print(f"Test MAPE: {mape:.2f}%")

    # Step 7: Save the model
    print("\n7. SAVING MODEL")
    print("-" * 40)
    model_path, scaler_path = save_model(model, scaler, train_losses, val_losses)

    print("\n" + "=" * 60)
    print("TRAINING COMPLETE!")
    print("=" * 60)

    return model, scaler, model_path, scaler_path

if __name__ == "__main__":
    # Run the main training function
    trained_model, fitted_scaler, model_path, scaler_path = main()

    print("\n" + "=" * 60)
    print("EXAMPLE USAGE FOR PREDICTION")
    print("=" * 60)
    print(f"""
    # To load and use the saved model:

    # Load the model
    import torch
    import joblib

    # Load model configuration and state dict
    checkpoint = torch.load('{model_path}')
    model = QuantumFinancialTCN(
        n_features=checkpoint['model_config']['n_features'],
        q_modes=checkpoint['model_config']['q_modes'],
        n_photons=checkpoint['model_config']['n_photons'],  # Will be 4
        hidden_channels=checkpoint['model_config']['hidden_channels'],
        kernel_size=checkpoint['model_config']['kernel_size'],
        dropout=checkpoint['model_config']['dropout']
    )
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()

    # Load the scaler
    scaler = joblib.load('{scaler_path}')

    # Make predictions on new data
    with torch.no_grad():
        # Prepare your data (same format as training)
        new_data = torch.FloatTensor(your_normalized_data).unsqueeze(0)  # [1, seq_len, n_features]
        predictions = model(new_data)  # [1, seq_len, n_features]
        next_step_pred = predictions[0, -1, :].numpy()

        # Inverse transform to get actual prices
        actual_prices = scaler.inverse_transform(next_step_pred.reshape(1, -1))
    """)
```
